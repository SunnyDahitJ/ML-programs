```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from sklearn import metrics
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Crop_recommendation.csv")

x = df.humidity

y = df.rainfall


x = [round(i,2) for i in list(x)]

y = [round(i,2) for i in list(y)]

coup = np.array(list(zip(x,y)))

print(coup)
distortions = []
K = range(1,8)
for i in K:
  print(i)
  model = KMeans(n_clusters=i)
  lan = model.fit_predict(coup)
  # print(lan)

  distortions.append(sum(np.min(cdist(coup, model.cluster_centers_,
'euclidean'), axis=1)) / coup.shape[0])
  # label = model.predict()


filtered_label0 = coup[lan == 0]
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.show()

u_labels = np.unique(lan)


for i in u_labels:
  plt.scatter(coup[lan==i,0], coup[lan==i,1],label=i)
plt.show()


# plt.xlabel("humidity")
# plt.ylabel("rainfall")
# plt.plot(x,y)

# plt.show()

# plt.plot(K, distortions, 'bx-')

# plt.show()

```
