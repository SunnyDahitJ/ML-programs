```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.cluster.hierarchy as shc  
from sklearn.model_selection import train_test_split 
import seaborn as sns
```

```python
df = pd.read_csv("Crop_recommendation.csv")

data = df.iloc[: , 3:5]
data
datas=data[:200]
```

```python
cluster = shc.linkage(datas, method="ward")
shc.dendrogram(Z=cluster)
# plt.title("Dendrogram Plot using ward")  
# plt.ylabel("Euclidean Distances")  
# plt.xlabel("indexs")  
plt.show()  
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage="ward")
model.fit(datas)
```

```python
data_labels = model.labels_
sns.scatterplot(x="humidity", y = "temperature", data= datas, hue=data_labels,palette=['red','blue','green']).set_title("Using Wards")
plt.show()
```

```python
cluster1 = shc.linkage(datas, method="single")
shc.dendrogram(Z=cluster1)
plt.show()
```

```python
model1= AgglomerativeClustering(n_clusters=2, linkage="single")
model1.fit(datas)

```

```python
data_labels = model1.labels_
sns.scatterplot(x= "humidity", y="temperature", data = datas, hue = data_labels, palette =['red', 'blue']).set_title("single linkage")
plt.show()
```
