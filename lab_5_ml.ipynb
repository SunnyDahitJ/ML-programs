```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.cluster.hierarchy as shc  
from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_csv("Crop_recommendation.csv")

feature1 = df["rainfall"]

# feat = feature[:10]

feature2 = df["humidity"]

feature = [round(i,2) for i in list(feature1)]

label = [round(i,2) for i in list(feature2)]


coup = np.array(list(zip(feature1,feature2)))

cop = coup[:10]

X , y = feature, label

# X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= .30 , shuffle =True)
```

```python
dendro = shc.dendrogram(shc.linkage(cop, method="ward"))  
plt.title("Dendrogrma Plot")  
plt.ylabel("Euclidean Distances")  
plt.xlabel("indexs")  
plt.show()  
```

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity="euclidean", linkage="ward")

y_pred= hc.fit_predict(coup)  
```

```python
plt.scatter(coup[y_pred == 0, 0], coup[y_pred == 0, 1],  c = 'blue', label = 'Cluster 1')  
plt.scatter(coup[y_pred == 1, 0], coup[y_pred == 1, 1],  c = 'green', label = 'Cluster 2')  
plt.scatter(coup[y_pred== 2, 0], coup[y_pred == 2, 1],  c = 'red', label = 'Cluster 3')  
plt.title('Clusters of crops')  
plt.xlabel('features( rainfall and humidity')  
plt.ylabel('crops')  
plt.legend()  
plt.show()  
```
